# transaction_date_base

```python
##################################################
# For membership_loyalty
##################################################
transactions['how_long_it_will_expire'] = [i.days for i in (transactions.membership_expire_date - transactions.transaction_date)]
transactions['is_how_long_equal_to_plan_days'] = [1 if h_l_e==p_d else 0 for h_l_e, p_d in transactions[['how_long_it_will_expire','payment_plan_days']].values]
transactions['is_how_long_longer_than_plan_days'] = [1 if h_l_e > p_d else 0 for h_l_e, p_d in transactions[['how_long_it_will_expire','payment_plan_days']].values]
transactions['days_longer_than_plan_days'] = [i if i > 0 else 0 for i in (transactions.how_long_it_will_expire - transactions.payment_plan_days)]

##################################################
# 到期日在交易日之前...
##################################################

transactions['is_early_expiration'] = [1 if i.days < 0 else 0 for i in (transactions.membership_expire_date - transactions.transaction_date)]
transactions['early_expiration_days'] = [-i.days if i.days < 0 else 0 for i in (transactions.membership_expire_date - transactions.transaction_date)]
```

```python
output_col = ['msno','how_long_it_will_expire','is_how_long_equal_to_plan_days',
              'is_how_long_longer_than_plan_days','days_longer_than_plan_days',
             'is_early_expiration','early_expiration_days'
             ]
transactions[output_col]
```

# Feature Engineering --> w= 0~2 + -1(test)

```python
##################################################
# Load  
##################################################
col = ['msno','transaction_date','membership_expire_date']
transactions = utils.read_multiple_csv('../../input/preprocessed_data/transactions', col)
transactions = transactions.head(n = 500)

input_col = ['msno','t-1_transaction_date','t-1_membership_expire_date','order_number']
tran_time_diff = utils.read_multiple_csv('../../input/preprocessed_data/transactions_time_diff', input_col)
tran_time_diff = tran_time_diff.dropna() 
tran_time_diff = tran_time_diff.head(n = 500)

##################################################
# Convert string to datetime format
##################################################
transactions['membership_expire_date']  = transactions.membership_expire_date.apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
transactions['transaction_date']  = transactions.transaction_date.apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
tran_time_diff['t-1_membership_expire_date']  = tran_time_diff['t-1_membership_expire_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
tran_time_diff['t-1_transaction_date']  = tran_time_diff['t-1_transaction_date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
T = 0
train = pd.read_csv('../../input/preprocessed_data/trainW-{0}.csv'.format(T))
df = pd.merge(train, transactions, on='msno', how='left') # 此時msno就是不重複
df = pd.merge(df, tran_time_diff, on='msno', how='left').drop_duplicates(['msno','order_number'])

df = df.dropna()
```

```python
df[df.msno == '++1G0wVY14Lp0VXak1ymLhPUdXPSFJVBnjWwzGxBKJs=']
```

```python
def days_since_the_last_expiration(x):
    # x: each row of dataframe, series
    # return 負的代表消費者提早續約在上次交易到期之前,正的代表他在上次到期後才續約,0代表剛好那天續約
    x['days_since_the_last_expiration'] = [i.days for i in (x.transaction_date - x['t-1_membership_expire_date'])]
    return x
def days_since_the_last_subscription(x):
    x['days_since_the_last_subscription'] = [i.days for i in (x.transaction_date - x['t-1_transaction_date'])]
    return x
def is_subscribe_early(x):
    x['is_subscribe_early'] = [1 if i <0 else 0 for i in x.days_since_the_last_expiration]
    return x
df = df.groupby('msno').apply(days_since_the_last_expiration)
df = df.groupby('msno').apply(days_since_the_last_subscription)
df = df.groupby('msno').apply(is_subscribe_early)
```

```python
output_col = ['msno','days_since_the_last_expiration','days_since_the_last_subscription','is_subscribe_early']
df[output_col]
```
